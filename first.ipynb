```python
# Set Operations
# union() --> to combine two sets and return a new set with unique elements from both sets
# intersection() --> to find the common elements between two sets and return a new set
```

```python
set_1 = {10, 20, 30, 40, 50}
set_2 = {30, 40, 50, 60, 70}
print(set_1.union(set_2))  # Output: {10, 20, 30, 40, 50, 60, 70}
print(set_1.intersection(set_2))  # Output: {30, 40, 50}
```

```python
print(set_1.difference(set_2))  # Output: {10, 20}
print(set_2.difference(set_1))  # Output: {60, 70}
print(set_1.symmetric_difference(set_2))  # Output: {10, 20, 60, 70}
print(set_2.symmetric_difference(set_1))  # Output: {10, 20, 60, 70}
```

```python
# Operators in Python
# Arithmetic Operators --> +, -, *, /, //, %, **
# Assignment Operators --> =, +=, -=, *=, /=, //=, %=
# Comparison Operators --> ==, !=, >, <, >=, <=
# Logical Operators --> and, or, not
# Identity Operators --> is, is not
# Membership Operators --> in, not in
# Bitwise Operators --> &, |, ^, ~, <<, >>
```

```python
# Arithmetic Operators
a = 10
b = 20
c = a + b
print(c)
c = a - b
print(c)
c = a * b
print(c)
c = a / b
print(c)
c = a // b
print(c)
c = a % b
print(c)
c = a ** b
print(c)
```

```python
# Assignment Operators
a = 10
a += 5
print(a)
a -= 5
print(a)
a *= 5
print(a)
a /= 5
print(a)
a //= 5
print(a)
a %= 5
print(a)
a **= 5
print(a)
```

```python
# Comparison Operators
a = 10
b = 20
c = a == b
print(c)
c = a != b
print(c)
c = a > b
print(c)
c = a < b
print(c)
c = a >= b
print(c)
c = a <= b
print(c)
```

```python
#  Logical Operators (and, or,not)
a = True
b = False
c = a and b
print(c)
c = a or b
print(c)
c = not a
print(c)
c = not b
print(c)
```

```python
# Identity Operator (is, not is)
a = 10
b = 10
c = a is b
print(c)
c = a is not b
print(c)
```

```python
# Membership Operators(in, not in)
list_sample = ['vivo', 'oppo', 'apple', 'samsung']
a = 'vivo'
c = a in list_sample
print(c)
c = a not in list_sample
print(c)
b = 'nokia'
c = b in list_sample
print(c)
c = b not in list_sample
print(c)
```

```python
# Bitwise Operators (&, |, ^, ~, <<, >>)
a = 10
b = 20
c = a & b
print(c)
c = a | b
print(c)
c = a ^ b
print(c)
c = ~a
print(c)
c = a << 2
print(c)
c = a >> 2
print(c)
```
