```python
# Conditional or Control Flow Statements --> based on the condition, the flow of the program will be changed
# if, elif, else
```

```python
# Basic syntax for if
# if condition:
#     statements
```

```python
# example
if 10>5:
    print("10 is greater than 5")
print("Condition Statement Completed")   
```

```python
if 4>5:
    print("10 is greater than 5")
print("Condition Statement Completed") 
```

```python
if 4>5:
    print("10 is greater than 5")
else:
    print("10 is not greater than 5")    
print("Condition Statement Completed") 
```

```python
# Elif --> else if
# if condition:
#     statements
# elif condition:
#     statements
# else:
#     statements
```

```python
# Example
score = int(input("Enter your score: "))
if score >= 80:
    print("You got A grade")
elif score >= 60 and score < 80:
    print("You got B grade")
else:
    print("You got C grade")
```

```python
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
operator = input("Enter the operator: ")
if operator == "+":
    print(num1 + num2)
elif operator == "-":
    print(num1 - num2)
elif operator == "*":
    print(num1 * num2)
elif operator == "/":
    print(num1 / num2)
else:
    print("Invalid operator")
```

```python
# Loops in Python --> to do something multiple times
# 1. for loop --> number of iterations is known, to perform sequence operations
# 2. while loop --> number of iterations is unknown, it runs till the condition is true
# 3. nested loop
```

```python
# For Loop --> to do something multiple times
# Syntax --> for <variable_name> in range(<start>, <end>, <step>):
# range(start, end, step)
# range(0, 10, 1)
```

```python
#Examle for For loop
for i in range(0,10,1):
    print(i)
    if i == 5:
        break
```

```python
#Examle for For loop
for i in range(0,10):
    print(i)
    
```

```python
# Example for For Loop else statement
for i in range(1, 10, 2):
    print(i)
else:
    print("Loop Completed Successfully")    
```

```python
# Coditional Statements in Loop  (break, pass, continue)
# Break --> it breaks the loop(stop the execution of the loop and exits from the loop)
# Continue --> it skips the current iteration and moves to the next iteration
# Pass --> it does nothing
```

```python
# Break
for i in range (0, 10):
    print(i)
    if i == 5:
        break

print("Loop Complted Successfully")    
```

```python
# Continue
for i in range(0,10):
    
    if i == 5:
        continue
    print(i)
    
print("Loop Completed Successfully")
```

```python
# Pass
for i in range(0,10):
    
    if i == 5:
        pass
    print(i)
    
print("Loop Completed Successfully")
```

```python
# Sequence Operations
# 1. List
# 2. Tuple
# 3. String
```

```python
languages = ["Python", "Java", "C", "C++", "C#"]
for i in languages:
    print("Iteration",i)
    print("Index of the item",languages.index(i))
    
print("Loop Completed Successfully")
```

```python
for i in range(len(languages)):
    print("Iteration",languages[i])
    print("Index of the item",i)
    
print("Loop Completed Successfully")
```

```python
#tuple 
breakfast = ('Idli', 'Dosa', 'Vada', 'Poha', 'Poori')
for i in breakfast:
    print("Iteration",i)
    print("Index of the item",breakfast.index(i))
    
print("Loop Completed Successfully")
```

```python
# String
name = "Pranesh Dharun"
for i in name:
    print("Iteration",i)
    print("Index of the item",name.index(i))
    
print("Loop Completed Successfully")
```

```python
# While Loop
# Syntax --> while <condition>:
#             <statement>
#             <increment/decrement>
```

```python
# while Loop
i = 0
while i < 10:
    print(i)
    i += 1  # i = i + 1
```

```python
i = 0
while i < 10:
    print(i)
    i += 1  # i = i + 1
else:
    print("Loop Completed Successfully")    
```

```python
# break, continue, pass
i = 0
while i < 10:
    
    if i == 5:
        break
    print(i)
    i += 1  # i = i + 1
else:
    print("Loop Completed Successfully")
```

```python
i = 0
while i < 10:
    
    i += 1  # i = i + 1
    if i == 5:
        continue
    print(i)
else:
    print("Loop Completed Successfully")
```
