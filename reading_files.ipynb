```python
with open("example1.txt","r") as file_obj:
    content = file_obj.read()
    print(content)
```

```python
with open("example1.txt","r") as file_obj:
    content = file_obj.readlines()
    print(content)
    print(type(content))
with open("example.txt","a") as file_obj1:
    for i in content:
        file_obj1.write(i)
```

```python
file_obj = open("example1.txt","r+")
content = file_obj.read()
print(content)
file_obj.write("Naruto")
file_obj.close()
```

```python
file_obj = open("sample.txt","w+")
content = file_obj.readlines()
print(content)
file_obj.close()
```

```python
with open("sample.txt","a+") as file_obj:
    content = file_obj.read()
    print(content)
    file_obj.write(" Naruto")
    file_obj.close()
with open("sample.txt","r+") as file_obj:
    content = file_obj.read()
    print(content)
    file_obj.close()
```
